***ATM Opration***

```python
def withdraw(bal,amt):
    bal=bal-amt
    return bal
```

```python
def deposite(bal,amt):
    bal=bal+amt
    return bal
```

```python
def transfer(bal,amt):
    bal=bal-amt
    return bal
```

```python
def balcheck(bal):
    print(f"your current balence is {bal}")
```

```python
def stmt(data):
    for i in data:
        print(i)
```
