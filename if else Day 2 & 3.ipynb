***4]What is a while loop?***

A while loop is used to repeat a block of code as long as a given condition is true.
The loop stops automatically when the condition becomes false.

Syntax (Python)
while condition:
    # code to repeat

Short Example
i = 1
while i <= 5:
    print(i)
    i = i + 1


Output:

1
2
3
4
5

Explanation

i = 1 ‚Üí starting value

while i <= 5 ‚Üí condition

print(i) ‚Üí prints the value

i = i + 1 ‚Üí increases i (very important, otherwise loop runs forever)

Very Simple Example
while True:
    print("Hello")
    break


This runs once because break stops the loop.

```python
#reverse number program using while loop condition
num=int(input("enter a number:"))
rev=0
while num>0:
    digit=num%10
    rev=rev*10+digit
    num//=10
print("reverse=",rev)
```

```python
#palindrome number program using while lopp,if,else
num=int(input("enter a number"))
temp=num
rev=0
while temp>0:
   
    rev=rev*10+temp%10 #module using
    temp//=10
if num==rev:
    print("palindrome number")
else:
    print("not a palindrome")
    
```

```python
#Armstrong number using while loop and if else condition 
num = int(input("Enter a number: "))
temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum = sum + digit ** 3
    temp = temp // 10

if sum == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
#magic number using while lopp and if-else condition
num = int(input("Enter a number: "))

while num > 9:
    sum = 0
    while num > 0:
        sum = sum + num % 10
        num = num // 10
    num = sum

if num == 1:
    print("Magic Number") #true condtion
else:
    print("Not a Magic Number") #false condition
```
